```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
from pprint import pprint

from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

#print(list(lat_lngs)[0])

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial" #Fahrenheit

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
print(query_url)
```

```python
weather_response = []
x = 0

print("Beginning Data Retrieval")
print("-----------------------------")

for city in cities:
    response = requests.get(query_url + cities[x])
    if response.status_code == 200:
            if x <= 50:
                print(f'Processing Record {x} of Set 1 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1
            elif x <=100:
                print(f'Processing Record {x-50} of Set 2 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1
            elif x <= 150:
                print(f'Processing Record {x-100} of Set 3 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1                
            elif x <= 200:
                print(f'Processing Record {x-150} of Set 4 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1                
            elif x <= 250:
                print(f'Processing Record {x-200} of Set 5 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1        
            elif x <= 300:
                print(f'Processing Record {x-250} of Set 6 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1                
            elif x <= 350:
                print(f'Processing Record {x-300} of Set 7 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1
            elif x <= 400:
                print(f'Processing Record {x-350} of Set 8 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1
            elif x <= 450:
                print(f'Processing Record {x-400} of Set 9 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1                
            elif x <= 500:
                print(f'Processing Record {x-450} of Set 10 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1                
            elif x <= 550:
                print(f'Processing Record {x-500} of Set 11 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1                
            elif x <= 600:
                print(f'Processing Record {x-550} of Set 12 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1                
            elif x <= 650:
                print(f'Processing Record {x-600} of Set 13 | {cities[x]}')
                weather_response.append(response.json())
                x = x + 1
    else:
        print(f'City not found. Skipping...')
        x = x + 1
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

```python
len(weather_response)
```

```python
cities_list = []
cloudiness_list = []
countries = []
date_list = []
humidity_list = []
latitude_list = []
longitude_list = []
max_temp_list = []
wind_speeds = []
i = 0

for response in weather_response:
    city_name = weather_response[i]["name"]
    cloudiness = weather_response[i]["clouds"]["all"]
    country = weather_response[i]["sys"]["country"]
    date = weather_response[i]["dt"]
    humidity = weather_response[i]["main"]["humidity"]
    latitude = weather_response[i]["coord"]["lat"]
    longitude = weather_response[i]["coord"]["lon"]
    max_temp = weather_response[i]["main"]["temp_max"]
    wind_speed = weather_response[i]["wind"]["speed"]
    
    cities_list.append(city_name)
    cloudiness_list.append(cloudiness)
    countries.append(country)
    date_list.append(date)
    humidity_list.append(humidity)
    latitude_list.append(latitude)
    longitude_list.append(longitude)
    max_temp_list.append(max_temp)
    wind_speeds.append(wind_speed)
    
    i = i + 1
```

```python
print(len(cities_list))
print(len(cloudiness_list))
print(len(countries))
print(len(date_list))
print(len(humidity_list))
print(len(latitude_list))
print(len(longitude_list))
print(len(max_temp_list))
print(len(wind_speeds))
```

```python
df = pd.DataFrame({"City" : cities_list,
                   "Cloudiness" : cloudiness_list,
                   "Country" : countries,
                   "Date" : date_list,
                   "Humidity" : humidity_list,
                   "lat" : latitude_list,
                   "lng" : longitude_list,
                   "Max Temp" : max_temp_list,
                   "Wind Speed" : wind_speeds
})
```
